# Data Analysis and Exploration


# Import Necessary Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.graph_objs as go
import os
import glob
```

# Load and Explore Dataset


```python
# Define paths (relative to notebook location: src/backend/train/)
data_dir = "../models/data"

# Get all CSV files in the data directory
csv_files = glob.glob(os.path.join(data_dir, "*.csv"))

if not csv_files:
    print(f"No CSV files found in {data_dir}")
    print("Please check the data directory path.")
else:
    print(f"Found {len(csv_files)} dataset(s):")
    for f in csv_files:
        print(f"  - {os.path.basename(f)}")
    
    # Load first dataset for exploration
    if csv_files:
        appl = pd.read_csv(csv_files[0])
        appl['Date'] = pd.to_datetime(appl['Date'])
        appl['year'] = appl['Date'].dt.year
        appl['month'] = appl['Date'].dt.month
        appl['Open Price Change'] = appl['Open'].diff()
        print(f"\nLoaded: {os.path.basename(csv_files[0])}")
        print(appl.head())
```

```python
# Display dataset information
appl.info()
```

# Visualize Volume Trend


```python
# Plot volume trend
plt.figure(figsize=(15, 6))
plt.plot(appl['Date'], appl['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Volume trend over time')
plt.show()
```
